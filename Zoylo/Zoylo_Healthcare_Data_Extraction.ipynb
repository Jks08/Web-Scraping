```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup as soup
from urllib.request import Request,urlopen
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.keys import Keys
#Seting up webdrive
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome('D:\Software\chromedriver_win32\chromedriver.exe')
```

```python
url="https://www.zoylo.com/medicines/otc/healthcare-products.html?"
```

```python
link_2=[]
for i in range(1,278):
    link_2.append(url+"p="+str(i))
```

```python
len(link_2)
```

```python
pname=[]
cost=[]

```

```python
for i in range(len(link_2)):
        #url="https://www.1mg.com/labs/new-delhi/category"
            driver.get(link_2[i])
            #time.sleep(0.05)
            res=driver.execute_script("return document.documentElement.outerHTML")
            #parses the HTML
            soup_1=soup(res,'lxml')
            b=soup_1.findAll("div",{"class":"product-item-info"})
            for j in range(len(b)):
                #name
                n=b[j].find("strong",{"class":"product name product-item-name"})
                if n==None:
                    pname.append("")
                else:
                    pname.append(n.text.strip())

                #cost
                p=b[j].find("span",{"class":"price"})
                if p==None:
                    cost.append("Not Available in page")
                else:
                    cost.append(p.text)
            print(i+1)
```

Data Conversion

```python
column=["Name","Cost"]
df1=pd.DataFrame(columns=column)
```

```python
df1
```

```python
df1["Name"]=pd.Series(pname)
df1["Cost"]=pd.Series(cost)
```
