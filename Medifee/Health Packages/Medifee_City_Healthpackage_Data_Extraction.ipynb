```python
testname=[]
labname=[]
cost=[]
city=[]
tests=[]
```

```python
ss_1=0
```

```python
for i in range(52,len(links)):
    driver.get(links[i])
    #driver.get("https://www.medifee.com/health-checkup/basic-health-checkup-in-bhopal/")
    #once the scroller has scrolled all the way extract the required data
    res=driver.execute_script("return document.documentElement.outerHTML")
    soup_1=soup(res,'lxml')
    
    #heading
    head_1=soup_1.find("h1").text
    
    if " in " in head_1:
        
        x=soup_1.findAll("div",{"class":"panel panel-default"})
        for j in range(len(x)):
            if "Discount Offers" in x[j].text:
                #gets the name and city
                head=head_1.split(" in ")
                testname.append(head[0])
                city.append(head[1])
    
                #labname and cost
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split("Rs.")
                a_1=a[0].split("by")
                a_2=a_1[len(a_1)-1].split("\nOffer Price:")
                labname.append(a_2[0])
                cost.append(a[len(a)-1])
    
                
            elif "Related Tests" in x[j].text:
                ss_1+=1
                
            #elif "Thyrocare (Online Order)" in x[j].text:
             #   #gets the name and city
              #  head=head_1.split(" in ")
              #  testname.append(head[0])
              #  city.append(head[1])
    
                #labname and cost
           #     word=x[j].find("p",{"class":"lead"}).text
            #    a=word.split("Rs.")
             #   a_1=a[0].split("by")
             #   a_2=a_1[len(a_1)-1].split("Offer Price:")
             #   labname.append(a_2[0])
                
             #   a_3=x[0].find("span",{"class":"text-success"}).text
             #   cost.append(a_3)
                
                #address
             #   address.append("Free home sample pickup in "+head[1])
                #lab link
             #   lablink.append("Not Available")
            
            elif "Offer Price:" in x[j].text: 
                #gets the name and city
                head=head_1.split(" in ")
                testname.append(head[0])
                city.append(head[1])
    
                #labname and cost
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split("Rs.")
                a_1=a[0].split("by")
                a_2=a_1[len(a_1)-1].split("Offer Price:")
                labname.append(a_2[0])
                cost.append(a[len(a)-1])
                
                #test info
                word_1=x[j].text
                word_2=word_1.split(".00")
                tests.append(word_2[len(word_2)-1])
    
    
            
            else:
                #gets the name and city
                head=head_1.split(" in ")
                testname.append(head[0])
                city.append(head[1])
    
                #labname and cost
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split("Price: Rs.")
                labname.append(a[0])
                cost.append(a[len(a)-1])
                
                #test info
                word_1=x[j].text
                word_2=word_1.split(".00")
                tests.append(word_2[len(word_2)-1])
    
    
    print(i+1)
```

```python
labname_1=[]
for k in range(len(labname)):
    word_3=labname[k]
    word_4=word_3.split("By ")
    labname_1.append(word_4[len(word_4)-1])
labname=labname_1
```

```python
test_1=[]
for k in range(len(tests)):
    word_5=""
    word_3=tests[k]
    word_4=word_3.split("\n")
    for s in range(len(word_4)):
        if s!=0:
            word_5=word_5+","+word_4[s]
        else:
            word_5+=word_4[s]
    test_1.append(word_5)
tests=test_1
```
