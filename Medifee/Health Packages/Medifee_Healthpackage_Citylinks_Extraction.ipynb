Scrolling Infinite Times

```python
df=pd.read_csv("medifeehealthpackagelinks.csv")
links=df["Link"]
```

Parsing the Page

```python
testname=[]
overview=[]
caution=[]
#faq=[]
```

```python
For Details
```

```python
for i in range(len(links)):
    ovr=""
    caut=""
    driver.get(links[i])
    #once the scroller has scrolled all the way extract the required data
    res=driver.execute_script("return document.documentElement.outerHTML")
    soup_1=soup(res,'lxml')
    
    if soup_1.find("h1")!=None:
        word=soup_1.find("h1").text
        a=word.split("Cost")
        testname.append(a[0])
        #gets the name and link
        c1=soup_1.findAll("p")
        for k in range(len(c1)-2):
            if c1[k].text=="" and c1[k+1].text=="":
                count=k
                count_1=k+2
            elif c1[k].text=="" and "Just select your city." in c1[k+1].text and c1[k+2].text=="":
                count=k
                count_1=k+3
        for r in range(1,count):
            ovr+=c1[r].text
        overview.append(ovr)
        
        for e in range(count_1,len(c1)-1):
            caut+=c1[e].text
        caution.append(caut)
        
        
        #FAQ
        #c2=soup_1.findAll("div",{"class":"panel-body"})
        #faq_1=[]
        #for j in range(len(c2)):
         #   if "Q: " in c2[j].text:
         #       faq_1.append(c2[j].text)
        #faq.append(faq_1)
    
    print(i+1)
```

```python
overview_1=[]
for j in range(len(overview)):
    x_1=""
    word=caution[j]
    a=word.split("\n\n")
    for s in range(len(a)):
        x_1+=a[s]
    overview_1.append(x_1)
```

Data Conversion

```python
column=["Test Name","Overview","Procedure and Precaution"]
```

```python
df1=pd.DataFrame(columns=column)
df1["Test Name"]=pd.Series(testname)
df1["Overview"]=pd.Series(overview)
df1["Procedure and Precaution"]=pd.Series(caution)
#df1["FAQ"]=pd.Series(faq)
```

For Links 

```python
testname=[]
link=[]
city=[]
avgprice=[]
maxprice=[]
minprice=[]
```

```python
for i in range(len(links)):
    driver.get(links[i])
    #once the scroller has scrolled all the way extract the required data
    res=driver.execute_script("return document.documentElement.outerHTML")
    soup_1=soup(res,'lxml')
    
    if soup_1.find("h1")!=None:
        word=soup_1.find("h1").text
        b=word.split("Cost")
        #gets the name and link
        c1=soup_1.findAll("tr")
        for j  in range(1,len(c1)):
            testname.append(b[0])
            word=c1[j].text
            a=word.split("Rs. ")
            if len(a)==4:
                city.append(a[0])
                avgprice.append(a[1])
                minprice.append(a[2])
                maxprice.append(a[3])
            else:
                city.append(a[0])
                avgprice.append(a[1])
            link_1=c1[j].find("a")
            link_2=link_1.get("href")
            link.append("https://www.medifee.com"+link_2)
                
    print(i+1)   
```
