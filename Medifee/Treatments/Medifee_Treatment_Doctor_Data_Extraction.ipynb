Parsing the Page

```python
doctorname=[]
specialisation=[]
experience=[]
study=[]
hospital=[]
about=[]
city=[]
```

```python
ss_1=0
```

```python
for i in range(len(links)):
    driver.get(links[i])
    time.sleep(0.5)
    #driver.get("https://www.medifee.com/tests/adenosine-deaminase-test-cost-in-ahmedabad/")
    #once the scroller has scrolled all the way extract the required data
    res=driver.execute_script("return document.documentElement.outerHTML")
    soup_1=soup(res,'lxml')
    
    #heading
    head_1=soup_1.find("h1").text
    
    if " in " in head_1:
        
        x=soup_1.findAll("div",{"class":"panel panel-default"})
        head=head_1.split(" in ")
        for j in range(len(x)):
            if "Various Indian Cities" in x[j].text:
                ss_1+=1
            else:
                #gets the doctor name and city
                city.append(head[1])
                name_1=x[j].find("h2")
                if name_1==None:
                    hos=x[j].find("p",{"class":"lead"})
                    word=hos.text
                    word_1=word.split("\n")
                    word_2=word_1[0].split("-")
                    doctorname.append(word_2[0])
                    
                    #experience,specialisation
                    word_3=word_1[1].split("-")
                    specialisation.append(word_3[0])
                    experience.append(word_3[1])
                    
                    #study
                    study_1=x[j].find("span",{"class":"text-success"})
                    study.append(study_1.text)
                    
                    #hospitals
                    a_1=x[j].text
                    a_2=a_1.split("Hospital(s):")
                    a_3=a_2[1]  
                    a_4=a_3.split("\n")
                    a_5=""
                    for a11 in range(len(a_4)):
                        if a_4[a11]!='':
                            a_5=a_5+a_4[a11]
                    hospital.append(a_5)
                        
                    #about
                    about.append("Not Available in page")
                else:
                    doctorname.append(name_1.text)
            
                #study
                    study_1=x[j].find("span",{"class":"text-success"})
                    study.append(study_1.text)
            
                #hospital,specialization and experience 
                    hos=x[j].find("p",{"class":"lead"})
                    word=hos.text
                    word_1=word.split("\n")
                    word_2=word_1[0].split("-")
                    specialisation.append(word_2[0])
                    experience.append(word_2[len(word_2)-1])
                    hospital.append(word_1[len(word_1)-1])
            
                #about
                    about_1=x[j].text
                    about_2=about_1.split("\n")
                    if about_2[len(about_2)-3]=="":
                        about.append(about_2[len(about_2)-4])
                    else:
                        about.append(about_2[len(about_2)-3])
            
    print(i+1)
```

```python
hospital_1=[]
for p_1 in range(len(hospital)):
    p_2=hospital[p_1]
    p_3=p_2.split("Hospital(s):")
    p_4=""
    for p_5 in range(len(p_3)):
        if p_3[p_5]!='':
            if p_5==len(p_3)-1:
                p_4=p_4+p_3[p_5]
            else:
                p_4=p_4+p_3[p_5]+";"
    hospital_1.append(p_4)
hospital=hospital_1
```

Data Conversion

```python
column=["Doctor Name","Specialization","Experience","Qualification","Hospital(s)","About","City"]
```

```python
df=pd.DataFrame(columns=column)
```

```python
df
```

```python
df["Doctor Name"]=pd.Series(doctorname)
df["Specialization"]=pd.Series(specialisation)
df["Experience"]=pd.Series(experience)
df["Qualification"]=pd.Series(study)
df["Hospital(s)"]=pd.Series(hospital)
df["About"]=pd.Series(about)
df["City"]=pd.Series(city)

```
