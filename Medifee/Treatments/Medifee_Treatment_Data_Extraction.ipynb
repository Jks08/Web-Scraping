Parsing the Page

```python
surgeryname=[]
hospitalname=[]
address=[]
cost=[]
city=[]

doctorsurgery=[]
doctorcity=[]
doctorlink=[]
```

```python
ss_1=0
```

```python
for i in range(300):
    driver.get(links[i])
    time.sleep(0.5)
    #driver.get("https://www.medifee.com/tests/adenosine-deaminase-test-cost-in-ahmedabad/")
    #once the scroller has scrolled all the way extract the required data
    res=driver.execute_script("return document.documentElement.outerHTML")
    soup_1=soup(res,'lxml')
    
    #heading
    head_1=soup_1.find("h1").text
    
    if " in " in head_1:
        
        x=soup_1.findAll("div",{"class":"panel panel-default"})
        head=head_1.split(" in ")
        for j in range(len(x)-3):
            if "Need Doctor Consultation?" in x[j].text:
                #gets the surgery name and city
                surgeryname.append(head[0])
                city.append(head[1])
    
                #hospital name
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split(" by ")
                hospitalname.append(a[1])
        
                #cost
                cost_1=x[j].findAll("div",{"class":"col-lg-12"})
                cost_2=cost_1[(len(cost_1)-1)]
                cost.append(cost_2.text)
            
                #address
                text_1=x[j].findAll("div",{"class":"col-lg-7"})
                text_2=text_1[len(text_1)-1].text
                text_3=text_2.split("Location:")
                text_4=text_3[len(text_3)-1]
                address.append(text_4+","+head[1])
            
            elif "Get This Offer Now in Email" in x[j].text:
                #gets the surgery name and city
                surgeryname.append(head[0])
                city.append(head[1])
    
                #hospital name
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split(" by ")
                hospitalname.append(a[1])
        
                #cost
                cost_1=x[j].find("span",{"class":"text-success"})
                #cost_2=cost_1[(len(cost_1)-1)]
                cost.append(cost_1.text)
            
                #address
                text_1=x[j].text
                text_2=text_1.split("Location:")
                text_3=text_2[len(text_2)-1].split("Get This Offer Now in Email")
                address.append(text_3[0]+","+head[1])
                   
            else:
                #gets the surgery name and city
                surgeryname.append(head[0])
                city.append(head[1])
    
                #hospital name
                word=x[j].find("p",{"class":"lead"}).text
                a=word.split(" by ")
                hospitalname.append(a[1])
        
                #cost
                cost_1=x[j].findAll("div",{"class":"col-lg-12"})
                cost_2=cost_1[(len(cost_1)-1)]
                cost.append(cost_2.text)
                
                #address
                text_1=x[j].findAll("div",{"class":"col-lg-7"})
                text_2=text_1[len(text_1)-1].text
                text_3=text_2.split("Location:")
                text_4=text_3[len(text_3)-1]
                address.append(text_4+","+head[1])
    
        #Doctor link
        doc_1=soup_1.find("ul",{"class":"nav nav-tabs nav-justified"})
        if doc_1==None:
            ss_1=ss_1+1
        else:
            doc_4=doc_1.findAll("li")
            if len(doc_1)==3:
                doc_2=doc_4[2].find("a")
                doc_3=doc_2.get("href")
            doctorsurgery.append(head[0])
            doctorcity.append(head[1])
            doctorlink.append("https://www.medifee.com"+doc_3)
    
    print(i+1)
```

```python
cost_1=[]
for p_1 in range(len(cost)):
    p_2=cost[p_1]
    p_3=p_2.split("\n")
    p_4=""
    for p_5 in range(len(p_3)):
        if p_3[p_5]!='':
            if p_5==len(p_3)-1:
                p_4=p_4+p_3[p_5]
            else:
                p_4=p_4+p_3[p_5]+";"
    cost_1.append(p_4)
```

```python
cost_11=[]
for l in range(len(cost_1)):
    s_1=len(cost_1[l])
    word_11=cost_1[l]
    word=""
    for s_2 in range(s_1-1):
        word=word+word_11[s_2]
    cost_11.append(word)
#cost=cost_11
```
