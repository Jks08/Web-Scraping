```python
import bs4
import requests
import csv
import pandas as pd
import re
```

```python
resp=requests.get("https://www.medifee.com/diseases-faq.php")
resp.status_code
```

```python
soup=bs4.BeautifulSoup(resp.content,"html5")
link=soup.find_all("li")
li=[]
title=[]
link_main=[]
text_main=[]
for i in range (14,len(link)-1):
    b=link[i].find("a")
    li.append(b)
#     print(b)

for j in li:
    title.append(j.text)
    link_main.append(j['href'])
    print(j['href'])
for k in link_main:
    print(k)
for j in title:
    print(j)
len(title)
len(link_main)
```
