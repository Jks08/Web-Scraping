```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.apollopharmacy.in/'
driver.get(url)
```

```python
names = driver.find_elements_by_class_name('jss159 ')
name=[]
for i in range(len(names)):
    name.append(names[i].text)
```

```python
price = driver.find_elements_by_class_name('jss153 ')
original_price=[]
discount_price=[]
for i in range(len(price)):
    price[i]=price[i].text
    m= price[i]
    c= m.count('₹')
    if(c!=1):
        m = m.split(')')      
        original_price.append((m[0][5:]))
        discount_price.append( (m[1][1:] ) )
    else:
        original_price.append((m[4:]))
        discount_price.append( '' )
```

```python
img_m = driver.find_elements_by_class_name('jss152')
for i in range(len(img_m)):
    img_t = img_m[i].find_element_by_tag_name('img')
    img_m[i] = img_t.get_attribute('src')
    
```

```python
link = driver.find_elements_by_class_name('jss151')
for i in range(len(link)):
    l = link[i].find_element_by_tag_name('a')
    l = l.get_attribute("href")
    link[i] = l
```

```python
Speciality=[]
for i in range(len(link)):
    Speciality.append('Adult-care')
dt10 =pd.DataFrame(list(zip(name, original_price,discount_price,img_m , link, Speciality )),
               columns =['Name', 'original_price','discount_price','img_m' , 'link', 'Speciality'])
```
